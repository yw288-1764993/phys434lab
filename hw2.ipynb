2.numerically (computer program)

```python
p = np.array([1/6] * 6)
sump = np.convolve(p, p)
plt.xlabel("sums")
plt.ylabel("probability")
plt.xticks(range(2, 13))
plt.ylim(0, 1/5)
plt.vlines(range(2, 13), 0, sump, linewidth = 20);
```

3.

```python
mean = sum(sump * np.arange(2, 13))
variance = sum(sump * (np.arange(2, 13) - mean)**2)
```

```python
def average(trials, width=5):
    avgp = np.copy(p)
    for i in range(trials):
        avgp = np.convolve(p, avgp)
    plt.xlabel("average of 10 dice")
    plt.ylabel("probability")
    pofavg = np.linspace(1, 6, len(avgp))
    plt.vlines(pofavg, 0, avgp, linewidth = 5);
    plt.xlim(1, 6)
    plt.ylim(0, None)
    return pofavg, avgp
average(10);
```

The average does follow a gaussian distribution. There are peak and wing pattern point it to a gaussian. It will be a better approximation when the sample size increase.

5.

```python
pofavg, avgp = average(100, width=2)
mean = sum(avgp * pofavg)
variance = sum(avgp * (pofavg - mean)**2)
ax = plt.gca()
plt.xlabel("average of 100 dice")
plt.xlim(2, 5)
```

It clearly follows the gaussian distribution, and it's a better approximation. When the sample size increase, the bias will decrease, independent random variables will also decrease.
