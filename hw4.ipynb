```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
matplotlib.rc('figure', figsize=(10, 6))
```

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
plt.scatter(t, bkgd, s=1)
plt.xlabel("time")
plt.ylabel("background")
```

Problem 1

1.a

```python
data_points = bkgd[:1000]
time_range = t[:1000]
plt.scatter(time_range, data_points)
plt.xlabel("time")
plt.ylabel("background")
```

1.b

```python
def plot_everywhere(time_range,data_points,start,size):
    data_points = data_points[start:(start+size)]
    time_range = time_range[start:(start+size)]
    plt.scatter(time_range,data_points,s=30)
    plt.xlabel("time")
    plt.ylabel("background")
plot_everywhere(t, bkgd, 500, 2000)
```

1.c

```python
def plot_sample(x,t,bkgd):
    plot_x = bkgd[np.arange(0,bkgd.shape[0],x)]
    plt.scatter(t[np.arange(0,bkgd.shape[0],x)],plot_x)
    plt.xlabel("time")
    plt.ylabel("background")
plot_sample(100,t,bkgd)
```
