```python
noise = stats.poisson.rvs(mu=100, size=100000)
injection = np.linspace(100,500,100000)
observed = noise + injection

h = plt.hist2d(observed, injection, bins = (50,50), density=True)
plt.colorbar(h[3])
plt.xlabel("Observed Data", fontsize = "20")
plt.ylabel("Injected Signal", fontsize = "20")
```

b)

If selecting an injection value of 190 on Injected Signal, we can see it corresponds to a Observed Data about 290. 190 + 100 = 290. It matched the answer in problem2.
